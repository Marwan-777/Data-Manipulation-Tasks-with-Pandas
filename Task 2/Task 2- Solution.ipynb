```python
import pandas as pd
import json


requests = [json.loads(line) for line in open("usa.gov_click_data.json")]
print('Number of requests: ', len(requests))
print()
display(requests[0])
```

```python
df = pd.DataFrame(columns = ['info','from_url','to_url','city','ll','timezone','time_in','time_out'])
row = []

for req in requests:
   row.append(req.get("a"))
   row.append(req.get("r"))
   row.append(req.get("u"))
   row.append(req.get("cy"))
   row.append(req.get("ll"))
   row.append(req.get("tz"))
   row.append(req.get("t"))
   row.append(req.get("hc"))
   df.loc[df.shape[0]] = row
   row = []

```

## Raw data is ready for some transformations

```python
df
```

## Extracting the OS and Browser names from the info column

```python
!pip install user_agents
from user_agents import parse
df["web_browser"] = df["info"].apply(lambda x: parse(x).browser.family)
df["operating_sys"] = df["info"].apply(lambda x: parse(x).os.family)
df.drop(["info"],axis =1, inplace = True)
df
```

## Extracting longitude and latitude

```python
df["longitude"] = df["ll"].apply(lambda x: x if x is None else x[0])
df["latitude"] = df["ll"].apply(lambda x: x if x is None else x[1])
df.drop(["ll"],axis = 1,inplace = True)
df.head()
```

## Converting time-in and time_out to UTC timestamp

```python
import datetime

df["time_in"] = df["time_in"].apply( datetime.datetime.utcfromtimestamp )
df["time_out"] = df["time_out"].apply( datetime.datetime.utcfromtimestamp )
df.rename(columns = {'time_in':'time_in (UTC)','time_out':'time_out (UTC)'}, inplace = True)
df.head()
```

## Getting the GMT timezone instead of its name

## Ex: 'America/New_York' will be 'GMT -05:00'


```python
import pytz
import datetime

def get_timezone(str):
  if str is None or str == '':
    return None
  timezone = pytz.timezone(str)
  gmttime = timezone.localize(datetime.datetime.now()).strftime('GMT %z')
  return gmttime
```

```python
df["timezone"] = df["timezone"].apply( get_timezone )
df.head()
```

## URL transformation

This function returns the domain name of the url specified


```python
from urllib.parse import urlparse
def get_short_url( long_url):

  parsed_url = urlparse(long_url)
  short_url = parsed_url.netloc
  if short_url == '':
    return None
  return short_url
```

```python
df["from_url"] = df["from_url"].apply ( get_short_url )
df["to_url"] = df["to_url"].apply( get_short_url)
df.head()
```

## Finally replacing any Null or None value with '...'

```python
df.fillna(value ="...",inplace = True)
df # Final dataframe
```
