# Step 10

Calculating the population by dividing the Energy Supply by the Energy Supply per Capita

which is the total energy supply and the energy supply per person thus, dividing them will give us the total number of people (population)

```python
ContinentDict  = {'China':'Asia', 
                  'United States':'North America', 
                  'Japan':'Asia', 
                  'United Kingdom':'Europe', 
                  'Russian Federation':'Europe', 
                  'Canada':'North America', 
                  'Germany':'Europe', 
                  'India':'Asia',
                  'France':'Europe', 
                  'South Korea':'Asia', 
                  'Italy':'Europe', 
                  'Spain':'Europe', 
                  'Iran':'Asia',
                  'Australia':'Australia', 
                  'Brazil':'South America'}


Joined_df["Continent"] = Joined_df.index.map(ContinentDict)
Joined_df["Population"] =Joined_df["Energy Supply"] / Joined_df["Energy Supply per Capita"]
Joined_df["Population"] = Joined_df["Population"].astype(np.int64)                         # rounding the result

```

```python
import numpy as np
df = Joined_df.groupby("Continent")["Population"].agg( Size = 'count' , Sum = 'sum', Mean = 'mean', Std = 'std')
display(df.head())
```
